# Koordinaten und Faceting mit ggplot2

Den Umgang mit Koordinaten zu lernen hilft uns dabei die Größe unserer Diagramme besser zu kontrollieren. Dabei erlaubt es uns das *Faceting* (Facettenschliff) mehrere Diagramme nebeneinander zu platzieren. Diese bezeihen sich für gewöhnich auf die selbe Datenquelle.

### Grunddiagramm und Daten

```python
library(ggplot2)
pl <- ggplot(mpg,aes(x=displ,y=hwy)) + geom_point()
```

```python
pl
```

## x und y Limits festlegen

Du kannst `scale_x_continuous` und `scale_y_continuous` mit einem zusätzlichen `limits=c(low,high)` Parameter übergeben, um den Maßstab anzupassen. Ein etwas schönerer Weg ist die Verwendung von `+ coord_cartesian()` mit `xlim` und `ylim` Parametern und numerischen Vektoren. Ein Beispiel:

```python
pl + coord_cartesian(xlim=c(1,4),ylim=c(15,30))
```

## Seitenverhältnis

WIr können `coord_fixed()` benutzen, um die Seitenverhältnisse unseres Diagramms (Standard ist 1:1) festzulegen:

```python
# Seitenverhältnis, ausgedrückt als y / x
pl + coord_fixed(ratio = 1/3)
```

# Facets

Der beste Weg um ein *Facet Grid*, also mehrere Diagramme, nebeneinander zu platzieren ist die Verwendung von `facet_grid()`:

```python
help(facet_grid)
```

```python
p <- ggplot(mpg, aes(displ, cty)) + geom_point()

p + facet_grid(. ~ cyl)
```

```python
p + facet_grid(drv ~ .)
```

```python
p + facet_grid(drv ~ cyl)
```

Die [Dokumentation](http://docs.ggplot2.org/current/facet_grid.html) verrät euch mehr über die weiteren Optionen!
