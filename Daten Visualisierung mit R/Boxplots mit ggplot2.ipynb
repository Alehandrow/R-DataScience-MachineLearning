# Boxplots mit ggplot2

[Boxplots](https://en.wikipedia.org/wiki/Box_plot) sind eine übliche Herangehensweise an die Darstellung von Gruppen numerischer Werte durch ihre Quartile. Boxplots können außerdem auch Linien (en. whiskers) haben, die aus den Boxen hervorstehen und das obere und untere Quartil darstellen. Daher ergibt sich die Bezeichnung "box-and-whisker plot". Ausreiser werden dabei als einzelne Punkte gezeigt.

```r
library(ggplot2)
df <- mtcars
```

```r
head(df)
```

## qplot

```r
qplot(factor(cyl), mpg, data = mtcars, geom = "boxplot")
```

## ggplot

```r
pl <- ggplot(mtcars, aes(factor(cyl), mpg))

pl + geom_boxplot()
```

```r
pl + geom_boxplot() + coord_flip()
```

```r
pl + geom_boxplot(aes(fill = factor(cyl)))
```

```r
pl + geom_boxplot(fill = "grey", color = "blue")
```
